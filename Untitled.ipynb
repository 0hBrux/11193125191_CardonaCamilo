```python
#%%         EJERCICIO 5
       

import keyword
LR = keyword.kwlist   
x = input('Digite un texto ')
VOM = ['A','E','I','O','U']
LT = ['Á','É','Í','Ó','Ú','á','é','í','ó','ú']      #utilizo listas para comparar
vm = 0
t = 0
d = 0
e = 0                      # EMPIEZO LOS CONTADORES EN 0 DE LO QUE ME PIDEN
r = 0

for n in range(LR):       # EN EL RANGO DE LAS PALABRAS SI HAY UNA CUENTELA
    if LR[n] in x:
        r = r + 1
print('En el texto que usted ingresó hay ',r,' palabras reservadas')

for i in VOM:
    for o in x:      # compara la lista de mayusculas con cada letra del texto
        if i==o:        # si la encuentra sumele una al contador
            vm = vm +1             
print('En el texto que usted ingresó hay ',vm,' vocales mayusculas')

for l in LT:                     # SUMELE UNA AL CONTADOR
    for j in x:
        if l==j:
            t = t + 1
print('En el texto que usted ingresó hay ',t, 'vocales con tildes') 
   
   
for n in x:                        
    if n.isdigit() == True:  #utiliza el comando .isdigit()para encontrar digitos
          d = d +1
          
    if n.isspace() == True:  #utiliza el comando .isspace() para encontrar espacios
          e = e + 1
      
print('En el texto que usted ingresó hay ',d,' numeros')
print('En el texto que usted ingresó hay ',e,' espacios')
```

```python
#%%         EJERCICIO  10

print('Este programa crea un vector con los elementos de una matriz, extrayendolos elemen\
tos de forma de espiral en sentido de las manecillas del reloj')

m = []   #Creamos vector vacío

filas = int(input('Ingrese el numero de filas de la matriz: '))
columnas = int(input('Ingrese el numero de columnas de la matriz: '))

for i in range(filas):
    m.append([])     #Añadimos el corchete para las filas de la matriz
    for j in range(columnas):
        m[i].append(int(input(f'Escriba el elemento #{j+1} de la fila {i+1}: \
         ')))

num_elementos = filas*columnas     #numero de elementos de la matriz    
vec = []                      #vector vacio para los elementos de la matriz
col_final = columnas + 1   #esta es la ultima columna de la matriz
col_inicial = 0               #Primera columna
fil_final = filas - 1           #ultima fila
fil_inicial = 0                #primera fila
cont = 0         #en este contador va la cantidad de elementos extraidos de la matriz

while cont < num_elementos:  #ciclio donde tiene limite en la cantidad de elementos extraidos
    for i in range(col_final,col_inicial,-1):  # recorridode iziquierda a derecha desde ultimo elemento de sublista
        pos = m[fil_final][i]                      #extraemos un elemento de la matriz
        vec.append(pos)                        #agregamos el elemento extraido al vector
        cont +=1                            #contador de la cantidad de elementos extraidos
        
    for i in range(fil_final,fil_inicial,-1):    #segundo recorrido
        pos = m[i][col_inicial]                 #extrae elemento
        vec.append(pos)                        #agrega a vector
        cont +=1                              #contador
        
    for i in range(col_inicial,col_final): #tercer recorrido
        pos = m[fil_inicial][i]           #extrae elemento
        vec.append(pos)                  #agrega a vector
        cont +=1                        #contador
        
    for i in range(fil_inicial,fil_final):  #cuarto recorrido
        pos = m[i][col_final]              #extrae elemento 
        vec.append(pos)                   #agrega a vector
        cont +=1                         #contador 
        
    col_final -= 1            
    col_inicial += 1        #se ajustan las posiciones
    fil_final -= 1         #de los proximos recorridos de los ciclos
    fil_inicial +=1
    
print(vec)
```

```python
#%%         EJERCICIO   12

x = int(input("Digite la altura: "))  #definimos altura del triangulo
c = "*"  #caracter que nos piden utilizar

def triangulo(x,c):   #funcion de altura y caracter
    for i in range(1,x+1): # en el primer recorrido va colocar el caracter 
        print(c*i)    # primero uno, luego dos veces luego tres hasta terminar uno mas de la altura
    for j in range(x-1,0,-1): # en el recorrido hace lo mismo pero empezando en uno menos de la altura
        print (c*j)    #hasta llegar a 1
     
    print()
    
triangulo(x,c) #llama la funcion
```
